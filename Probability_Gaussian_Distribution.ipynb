#Q_Q Plot

```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

import pylab
y=np.random.normal(3,5,size=100)
stats.probplot(y,dist='norm',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.uniform(low=1,high=100,size=1000)
stats.probplot(y_uni,dist='norm',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.uniform(low=1,high=100,size=1000)
stats.probplot(y_uni,dist='uniform',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.binomial(1,0.9,size=1000)
stats.probplot(y_uni,dist='uniform',plot=pylab)
pylab.show()
```

#Gaussian Distribution Analysis

```python
random_variable_population=np.random.normal(2,2,size=10)
counts,bins_edges=np.histogram(random_variable_population,bins=10,density=True)
pdf=random_variable_population/sum(random_variable_population)
cdf=np.cumsum(pdf)
plt.plot(bins_edges[1:],pdf)
plt.plot(bins_edges[1:],cdf)
plt.show()
```
