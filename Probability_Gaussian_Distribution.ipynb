#Q_Q Plot

```python
import scipy.stats as stats
import numpy as np
import pylab
y=np.random.normal(3,5,size=100)
stats.probplot(y,dist='norm',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.uniform(low=1,high=100,size=1000)
stats.probplot(y_uni,dist='norm',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.uniform(low=1,high=100,size=1000)
stats.probplot(y_uni,dist='uniform',plot=pylab)
pylab.show()
```

```python
y_uni=np.random.binomial(1,0.9,size=1000)
stats.probplot(y_uni,dist='uniform',plot=pylab)
pylab.show()
```
