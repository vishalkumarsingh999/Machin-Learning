# CDF Cumulative Distribution Functin 

```python
counts,bin_edges=np.histogram(iris_setosa['petal_length'],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
print(pdf)
print(cdf)
plt.title('PDF Vs CDF')

plt.plot(bin_edges[1:],pdf)
plt.legend(['PDF'])

plt.plot(bin_edges[1:],cdf,'r')

plt.show()
```

Repeating Above Plot for every iris flower fro petal length as we have analysis from PDF univariable Analysis

```python
#for setosa
counts,bin_edges=np.histogram(iris_setosa['petal_length'],density=True,bins=10) #default bins is 10 so it's optional
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


#for Versicolor
counts,bin_edges=np.histogram(iris_versicolor['petal_length'],density=True,bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#for verginica 
counts,bin_edges=np.histogram(iris_verginica['petal_length'],density=True,bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

# Mean,Variance and Standared Deviatation

```python
mean1=np.mean(iris_setosa['petal_length'])
mean2=np.mean(iris_verginica['petal_length'])
mean3=np.mean(iris_versicolor['petal_length'])
print('Petal_length Mean')
print('Setosa mean: ',mean1)
print('Verginica mean: ',mean2)
print('Versicolor: ',mean3)

std1=np.std(iris_setosa['petal_length'])
std2=np.std(iris_verginica['petal_length'])
std3=np.std(iris_versicolor['petal_length'])
print('\n\nPetal Length Standard Devitation')
print('Setosa: ',std1)
print('Verginica:',std2)
print('Vericolor:',std3)
```

# Median and Percentile

```python
median1=np.median(iris_setosa['petal_length'])
median2=np.median(iris_verginica['petal_length'])
median3=np.median(iris_versicolor['petal_length'])
print('\n\nPetal Length  Median')
print('Setosa: ',median1)
print('Verginica:',median2)
print('Vericolor:',median3)

percentile1=np.percentile(iris_setosa['petal_length'],q=95)
percentile2=np.percentile(iris_verginica['petal_length'],q=95)
percentile3=np.percentile(iris_versicolor['petal_length'],q=95)
print('\n\nPetal Length 95 Percentile')
print('Setosa: ',percentile1)
print('Verginica:',percentile2)
print('Vericolor:',percentile3)

quantile1=np.percentile(iris_setosa['petal_length'],np.arange(0,100,25))
quantile2=np.percentile(iris_verginica['petal_length'],np.arange(0,100,25))
quantile3=np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25))
print('\n\nPetal Length 25,75,100 Quantiles')
print('Setosa: ',quantile1)
print('Verginica:',quantile2)
print('Vericolor:',quantile3)
```

```python
from statsmodels import robust
mad1=robust.mad(iris_setosa['petal_length'])
mad2=robust.mad(iris_verginica['petal_length'])
mad3=robust.mad(iris_versicolor['petal_length'])
print('\n\nPetal Length Median Absolute Deviation')
print('Setosa: ',mad1)
print('Verginica:',mad2)
print('Vericolor:',mad3)

quantile1=np.percentile(iris_setosa['petal_length'],np.arange(0,100,25))
quantile2=np.percentile(iris_verginica['petal_length'],np.arange(0,100,25))
quantile3=np.percentile(iris_versicolor['petal_length'],np.arange(0,100,25))
print('\n\nPetal Length Inter Quantile Range (75 percentile - 25 percentile)')
print('Setosa: ',quantile1[3]-quantile1[1])
print('Verginica:',quantile2[3]-quantile2[1])
print('Vericolor:',quantile3[3]-quantile3[1])
```
