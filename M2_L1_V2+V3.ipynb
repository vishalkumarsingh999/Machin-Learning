### pairplot of **all data set** of all feature with respect to other

It is way of visualization of n-D data.

```python
import seaborn as sns
data=sns.load_dataset('iris')
print(data.head())
sns.set_style('whitegrid')
sns.pairplot(data,hue='species',height=5,kind='scatter')
```

### Here is way to visualisation of 4D data as 4th column as color

```python
import plotly.express as px
data=px.data.iris()
print(data.head())

#4d plot is 3d plot with 4th dimination as color in 3d scatter plot

fig1=px.scatter_3d(data,x='sepal_length',y='sepal_width',z='petal_length',color='petal_width')
fig1.show()
```

```python
#now same plot with spicies as three different symbols/shapes
fig=px.scatter_3d(data,x='sepal_length',y='sepal_width',z='petal_length',color='petal_width',symbol='species')
fig.show()
```

```python
#more slyled 3d plot
fig2=px.scatter_3d(data,x='sepal_length',y='sepal_width',z='petal_length',color='petal_width',symbol='species',size='sepal_length',opacity=0.7)
fig2.show()
```
