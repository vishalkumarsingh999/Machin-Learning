# t-SNE on MNIST Dataset

## Dataset preprocssing and Standrization

```python
# -*- coding: utf-8 -*-
"""
Created on Thu Sep 16 17:06:54 2021

@author: Vishal Kumar Singh
"""


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data= pd.read_csv("/content/sample_data/mnist_train_small.csv")
x=data.iloc[:,1:]
y=data.iloc[:,[0]]

from sklearn.preprocessing import StandardScaler
std_x=StandardScaler().fit_transform(x)
```

## For perplexity =50 and iteration =3000

```python

from sklearn.manifold  import TSNE
model1=TSNE(n_components=2,random_state=0,perplexity=50,n_iter=3000)
tsne_x=model1.fit_transform(std_x)

import seaborn as sns
tsne_data=np.hstack((tsne_x,y))

tsne_df=pd.DataFrame(tsne_data,columns=("Dim 1","Dim 2","Label"))

sns.FacetGrid(tsne_df,hue="Label").map(plt.scatter,"Dim 1","Dim 2").add_legend()
plt.show()
```

## For perplexity =100 and iteration =3000

```python
from sklearn.manifold  import TSNE
model1=TSNE(n_components=2,random_state=0,perplexity=100,n_iter=3000)
tsne_x=model1.fit_transform(std_x)

import seaborn as sns
tsne_data=np.hstack((tsne_x,y))

tsne_df=pd.DataFrame(tsne_data,columns=("Dim 1","Dim 2","Label"))

sns.FacetGrid(tsne_df,hue="Label").map(plt.scatter,"Dim 1","Dim 2").add_legend()
plt.show()
```

## For perplexity = 100 and iteration = 5000

```python
from sklearn.manifold  import TSNE
model1=TSNE(n_components=2,random_state=0,perplexity=100,n_iter=5000)
tsne_x=model1.fit_transform(std_x)

import seaborn as sns
tsne_data=np.hstack((tsne_x,y))

tsne_df=pd.DataFrame(tsne_data,columns=("Dim 1","Dim 2","Label"))

sns.FacetGrid(tsne_df,hue="Label").map(plt.scatter,"Dim 1","Dim 2").add_legend()
plt.show()
```

## For perplexity = 100 and iteration =5000 for 3D scatterplot

```python
from sklearn.manifold  import TSNE
model1=TSNE(n_components=3,random_state=0,perplexity=100,n_iter=3000)
tsne_x=model1.fit_transform(std_x)

import seaborn as sns
tsne_data=np.hstack((tsne_x,y))

tsne_df=pd.DataFrame(tsne_data,columns=("Dim 1","Dim 2","Dim 3","Label"))

sns.FacetGrid(tsne_df,hue="Label").map(plt.scatter,"Dim 1","Dim 2").add_legend()
plt.show()

import plotly.express as px
fig=px.scatter_3d(tsne_df,x="Dim 1",y="Dim 2",z="Dim 3",color="Label",opacity=0.8)
fig.show()
```
